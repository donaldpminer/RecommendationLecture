```python
# Read in the movies.csv file that has a lookup table id->movie name
movies = pandas.io.parsers.read_csv('ml-20m/movies.csv', index_col="movieId")[['title']]
movies[:5]
```

```python
# Helper function that maps id->movie name (useful later)
def id2title(idnum):
    return movies.loc[idnum][0]

print id2title(2)
print id2title(150)
```

```python
# Load up the ratings data. This is the meat. I drop timestamp because I need memory
#   more than I need timestamps.
ratings = pandas.io.parsers.read_csv('ml-20m/ratings.csv')[['movieId', 'userId', 'rating']]
ratings[:5]
```

```python
# LET'S DO SOME DATA EXPLORATION

# Number of ratings per user
users = ratings.groupby('userId')
usersc = users.count()
print 'number of ratings', len(ratings)
print 'number of users', len(usersc)
print 'mean', usersc.mean()[0]
print 'median', usersc.median()[0]
print 'min', usersc.min()[0]
print 'max', usersc.max()[0]
```

```python
# Number of ratings per movie
movie_ratings = ratings.groupby('movieId')
movie_ratingsc = movie_ratings.count()
print 'number of movies', len(movie_ratingsc)
print 'mean', movie_ratingsc.mean()[0]
print 'median', movie_ratingsc.median()[0]
print 'min', movie_ratingsc.min()[0]
print 'max', movie_ratingsc.max()[0]
```

```python
# LET'S IMPLEMENT THE DISTANCE METRIC NOW

# We're going to use Euclidean here because of simplicity
#  There are others...
Image('euclidean_pearson_similarity.png')
```

```python
# This function compares the movies that the two users have seen
def compare_users(u1, u2):
    u1 = users.get_group(u1)
    u2 = users.get_group(u2)
    
    rs = pandas.merge(u1, u2, on=['movieId'], how="inner")

    return rs

compare_users(1,5)[:10]
```

```python
# Calculates the Euclidean distance between two users
def euclidean(u1, u2):
    c = compare_users(u1, u2)
    
    # compute the sqroot of the sum of squares
    return sum((c['rating_x'] - c['rating_y'])**2)**.5, len(c)
    

euclidean(1,5)
```

```python
# Tanimoto is a different distance metric for binary attributes (instead of scores)

# It is the number of items they share in common divided by the total (not union) number of items
#    they have both selected.

Image("tanimoto_koeffizient.png")

```

```python
USERS = ratings['userId'].unique()

# Finds the closest 'k' users to 'u' using the 'd' distance function
def knn(u, k, d):
    topk = []

    # For each other user...
    for u2 in USERS: # If I want to run overnight...
#    for u2 in random.sample(USERS, 10000): # I'm doing a SAMPLE beause my laptop sux

        if u - u2 == 0: continue  # I'm the same user!

        # Figure out how far away they are from one another
        distance, count = d(u, u2)
        
        # The distance isn't very interesting if they don't have many movies in common
        if count > 5:
            topk.append((distance, count, u2))

    # pull the top K        
    topk.sort()
    topk = topk[:k]
    
    return topk

topk1 = knn(1, 120, euclidean)
topk1[:10]
```

```python
# Finds the movies u2 has seen that u1 has not
#    This function is useful to pull the recommendations
def userdiff(u1, u2):
    u1 = users.get_group(u1)
    u2 = users.get_group(u2)
    
    # Implement a right outer join, then drop the intersection (super outer?)
    rs = pandas.merge(u1, u2, on=['movieId'], how="right")
    rs = rs[rs['rating_x'].map(math.isnan)][['movieId', 'rating_y']]
    
    return rs

userdiff(1, topk1[0][2])[:5]
```

```python
ms = {}

# Go through each of the top K and append the distances along with their ratings
for distance, count, u2 in topk1:
    for movieid, rating in userdiff(1, u2).get_values():
        if not ms.has_key(movieid):
            ms[movieid] = []
        
        ms[movieid].append((distance**-1, rating))

# Go through the aggregated distance&ratings and do a weighted sum on each
predictions = []
for movieid, values in ms.items():    
    # The total distance of the neighbors on this movie
    #    ... useful for weighted sums
    total_distance = sum( v[0] for v in values )

    # Perform the weighted sum (closer neighbors have more say)
    prediction = sum( v[1] * (v[0] / total_distance) for v in values )
    
    predictions.append(\
                (prediction, \
                 movie_ratings.get_group(movieid).median()[0], \
                 movie_ratings.get_group(movieid).mean()[0], \
                 id2title(int(movieid)), \
                 len(values) )  )

# Sort it so top scoring ones are first
predictions.sort(reverse=True)

o = pandas.DataFrame(predictions, columns=['Predicted', 'Median', 'Mean', 'Title', 'Neighbors'])
```
